# Part 3 Association Rules

### Driver Program For The Apriori Algorithm

```python
def apriori(data,minsup):
    
    count = 2
    candidate_one = make_one_candidate(data) # Create 1 item candidate
    transaction_sets = list(map(set, data)) # Make a set for each transaction
    frequent_one, support = prune(transaction_sets, candidate_one,minsup) # Get frequent 1 item set and supporting data by pruning
    frequent = [frequent_one] # List of frequent itemsets, starts off with frequent 1 itemset
 
    
    while(len(frequent[count - 2]) > 0):
        
        candidate_multiple = make_multiple_candidates(frequent[count-2], count) # Generate itemsets using F_k-1 x F_k-1
        frequent_multiple, support_multiple = prune(transaction_sets, candidate_multiple,minsup) # Prune itemsets
        support.update(support_multiple) # Update the support count
        frequent.append(frequent_multiple)
        count += 1
        
    return frequent, support   
```

### Function To Generate Frequent 1 itemsets

```python
# Make 1 itemset candidates
def make_one_candidate(data):
    
    candidates = [] # All items in transactions
    
    # Iterate through the transactions and make 1 itemset candidates 
    for transaction in data:
        
        for item in transaction:
        
            if not [item] in candidates:
            
                candidates.append([item])
    
    candidates.sort()


    return [set(x) for x in candidates]
```

### Function To Generate iItemsets With More Than 1 Item

```python
# Make itemsets that contain multiple items
def make_multiple_candidates(data, count):
    
    candidates = []
    size = len(data)
    
    # K-2 method to join sets
    for i in range(size):
        
        for j in range(i+1, size):
        
            L1 = list(data[i])[:count-2]
            L2 = list(data[j])[:count-2]
            L1.sort()
            L2.sort()
            
            if L1==L2:
            
                candidates.append(data[i] | data[j])
    
    return candidates
    
```

### Pruning Function

```python
# Prune and return valid itemsets
def prune(data, candidates,minsup):
    
    support_count = {}
    
    # For each transaction set, check if the candidate set is a subset
    # If so then update its support count
    for transaction in data:
        
        transaction_set = set(transaction)
        
        for candidate in candidates:
            
            if candidate.issubset(transaction_set):
                
                can = frozenset(candidate)
                
                if can not in support_count:
                    
                    support_count[can] = 1
                
                else:
                
                    support_count[can] += 1
    
    num_items = float(len(data))
    unpruned_candidates = []
    support_values = {}
    
    # Calculate support for every itemset
    for key in support_count:
        
        sup = 0
        sup = support_count[key]/num_items
        
        # If the support meets the minimum support requirements add it to the list of itemsets.
        if sup >= minsup:
        
            unpruned_candidates.insert(0, key)
            
        
        support_values[key] = sup
    
    return unpruned_candidates, support_values
    
```

### Function To generate rules

```python
def make_rules(items, data, minconf):
    
    ruleset = []
    size = len(items)
    
    # Go through each itemset 
    for i in range(1, len(items)):
    
        for freq in items[i]:
            
            itemset = [frozenset([item]) for item in freq] 
            
            if (i > 1):
                
                make_rules_aux(freq, itemset, data, ruleset, minconf) # Use the helper function for 2+ itemsets to generate rules
            
            else:
            
                confidence(freq, itemset, data, ruleset, minconf) # For 1 itemset just calculate the confidence

    return ruleset
```

### Helper Function For Rules

```python
def make_rules_aux(freq, itemset, data, ruleset, minconf):

    count = len(itemset[0]) 
    
    if count == 1: # 1  itemset no need to recursively check further
       
        items = confidence(freq, itemset, data, ruleset, minconf)
    
    if (len(freq) > (count+1)):
        
        items = make_multiple_candidates(itemset, count+1) 
        items = confidence(freq, items, data, ruleset, minconf) # Calculate Conf and retrieve X,Y,Sup from the confidence function
        
        if len(items) > 1: # Recursively check for large confidence values
            
            # See if there is iterative solution to this
            make_rules_aux(freq, items, data, ruleset, minconf)
```

### Calculate Confidence Funtion

```python
def confidence(freq, itemset, data, ruleset, minconf):
    
    rules = [] # List above the minimum confidence threshold
    
    for consequent in itemset: # Frequent itemsets
        
        denominator = freq-consequent
        conf = data[freq] / data[denominator] # Confidence = XY/X
        
        if conf >= minconf:
            
            ruleset.append((denominator, consequent, conf,data.get(freq))) # Adds X,Y,Conf,Sup to the ruleset 
            rules.append(consequent)

    return rules
```

### Isolate Frequent Itemsets From Support Data

```python
def isolate(data):
    
    return dict((k,v) for k,v in data.items() if v >= 0.025) # Only return those who meet the minsup
```

### Function To Print Support

```python
def print_support(data, mapDict):
    
    print("Frequent Itemsets and Their Support")
    print()

    for key, value in data.items():

        key = list(map(int,key)) # Convert string list to int list to access from dictionary

        #Getting the product name from the dictionary
        for number in range(len(key)):

            if key[number] in mapDict: # Find the number in product dictionary

                key[number] = mapDict[key[number]] # Replace number with appropriate name

        # Making it look nice
        key = ", ".join(repr(e) for e in key) # Change list to string
        key = key.replace('[','') # Remove brackets
        key = key.replace(']','')
        key = key.replace("'","") # Remove quotes
        value = round(value,4)
        print("{{{}}}    Support: {}".format(key, value))
```

### Function To Parse And Print Rules

```python
def print_rules(ruleset, mapDict):
    
    #Make it look nice

    #Remove the set classifiers
    for i in range(len(ruleset)):

        row = list(ruleset[i]) # Change from tuple to list
        
        for j in range(len(row)-2): # Ignore confidence and support

            col = list(map(int,row[j])) # String list to int list
            row[j] = col # Replacing the string list with int list

        ruleset[i] = row # Replacing the whole row with properly formatted data

    #Get values from dictionary
    for i in range(len(ruleset)):

        row = ruleset[i]

        for j in range(len(row)-2): # Ignore confidence and support

            col = row[j]

            for num in range(len(col)): # Converting from numbers to product names

                if col[num] in mapDict: # Find the number in product dictionary

                    col[num] = mapDict[col[num]] # Replace number with appropriate name

    #Pretty Printing
    print("Rules")
    print()
    for rule in ruleset:
        
        # Left Side
        x = rule[0]
        x = ", ".join(repr(e) for e in x) # Change list to string
        x = x.replace('[','') # Remove brackets
        x = x.replace(']','')
        x = x.replace("'","") # Remove quotes
        
        #Right Side
        y = rule[1]
        y = ", ".join(repr(e) for e in y)
        y = y.replace('[','')
        y = y.replace(']','')
        y = y.replace("'","") 
        
        # Confidence and Support
        conf = rule[2]
        sup = rule[3]
        
        # Round to 4 digits
        conf = round(conf,4)
        sup = round(sup,4)
        
        print("{{{}}} ---->  {{{}}}, Confidence = {}, Support: {}".format(x,y, conf, sup)) 
```

## 1000 Transactions

```python
Transactions = []
mapList = []
infile = open("1000/1000-out1.csv")
mapfile = open("products.csv")

for line in infile: # Input Parsing
    
    Transactions.append(line.strip().split(', ')[1:])

# Map the product descriptions to a dictionary

for line in mapfile: 
    
    mapList.append(line.strip().split(',')[1:])

mapDict ={}

for i in range(50):
    
    if i not in mapDict.keys():
        
        mapDict[i] = mapList[i]

mapfile.close()

#Running the apriori algorithm

#Itemsets contains a list of frequent itemsets
#Data is a dictionary that contains support values for all itemsets
itemsets, data = apriori(Transactions,0.025)

final = isolate(data) # Valid itemsets are extracted from the support data information

# Print the  final frequent itemsets and their support
print_support(final, mapDict)
```

```python
# Get the rules and print
ruleset = make_rules(itemsets,data,0.5)
print_rules(ruleset,mapDict)
infile.close()
```
